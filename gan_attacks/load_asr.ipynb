```python
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load GMI ASR

## Unprot GMI ASR

```python
distances_path = "asr/unprot_d_gmi"
distances_unprot_gmi = np.array(os.path.join(distances_path, "distances.npy"))
threshold=1.04
accuracy = np.mean(distances_unprot_gmi < threshold)
print(accuracy)
```

## MRP GMI ASR

```python
distances_path = "asr/mrp_d_gmi"
distances_mrp_gmi = np.array(os.path.join(distances_path, "distances.npy"))
threshold=1.04
accuracy = np.mean(distances_mrp_gmi < threshold)
print(accuracy)
```

## FE GMI ASR

```python
distances_path = "asr/fe_d_gmi"
distances_fe_gmi = np.array(os.path.join(distances_path, "distances.npy"))
threshold=1.04
accuracy = np.mean(distances_fe_gmi < threshold)
print(accuracy)
```

## GMI plot

```python
metric = "euclidean_l2"

sns.kdeplot(distances_mrp_gmi, label='MRP')
sns.kdeplot(distances_fe_gmi, label='FE')
sns.kdeplot(distances_unprot_gmi, label='Unprot')
plt.axvline(x=threshold, color='r', linestyle='--')
plt.xlabel('{} Distance'.format(metric))
plt.ylabel('Density')
plt.show()
```
