```python
import torch, os, time
import numpy as np
import torch.nn as nn
import sys
sys.path.append('../BiDO')
import model, utils, engine
```

```python
from model import InceptionResnetV1, InceptionResnetV1MRP, InceptionResnetV1FE
```

```python
eval_path = "../BiDO/eval_ckp"
eval_path_gmi = "../GMI/eval_model"
target_path = "../BiDO/target_model"
```

## Save eval model

```python
T_pytorch = InceptionResnetV1(pretrained="vggface2")
T_pytorch.eval()
```

```python
T_pytorch = nn.DataParallel(T_pytorch).cuda()

save_path = os.path.join(eval_path, "FaceNet_pytorch.tar")
torch.save({'state_dict': T_pytorch.state_dict()}, save_path)

save_path = os.path.join(eval_path_gmi, "FaceNet_pytorch.tar")
torch.save({'state_dict': T_pytorch.state_dict()}, save_path)
```

## Save (untrained) target model

### Unprotected 512-dim embeddings

```python
T_classify = InceptionResnetV1(pretrained="vggface2", classify=True, num_classes=1000)
T_classify.eval()
```

```python
save_path = os.path.join(target_path, "FaceNet_pytorch_classify_untrained.tar")
T_classify = nn.DataParallel(T_classify).cuda()
torch.save({'state_dict': T_classify.state_dict()}, save_path)
```

```python
save_path = os.path.join(target_path, "FaceNet_pytorch_classify_mrp_untrained.tar")
T_pytorch = nn.DataParallel(T_pytorch).cuda()
torch.save({'state_dict': T_pytorch.state_dict()}, save_path)
```
